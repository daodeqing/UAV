```python
import pandas as pd

# === Nom du fichier à vérifier ===
filename = "simulation_scenario.csv"

def verifier_conformite_csv(filename):
    print(f"🔍 Vérification de la conformité du fichier : {filename}")

    try:
        # Lecture brute
        with open(filename, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        if not lines:
            print("❌ Le fichier est vide.")
            return

        # --- 1️⃣ Détection du séparateur ---
        first_line = lines[0].strip('\n\r')
        sep = ',' if first_line.count(',') >= first_line.count(';') else ';'
        print(f"✅ Séparateur détecté : '{sep}'")

        headers = [h for h in first_line.split(sep)]
        nb_cols = len(headers)
        print(f"✅ Nombre de colonnes attendu : {nb_cols}")
        print(f"→ En-têtes trouvés : {headers}")

        # --- 2️⃣ Vérification précise des espaces dans les noms de colonnes ---
        cols_with_leading_spaces = [h for h in headers if h.startswith(' ')]
        cols_with_trailing_spaces = [h for h in headers if h.endswith(' ')]
        cols_with_inner_spaces = [h for h in headers if ' ' in h.strip()]

        if cols_with_leading_spaces:
            print(f"⚠️ Espaces **au début** des noms de colonnes : {cols_with_leading_spaces}")
        if cols_with_trailing_spaces:
            print(f"⚠️ Espaces **à la fin** des noms de colonnes : {cols_with_trailing_spaces}")
        if cols_with_inner_spaces:
            print(f"⚠️ Espaces **internes** (au milieu) : {cols_with_inner_spaces}")

        if not (cols_with_leading_spaces or cols_with_trailing_spaces or cols_with_inner_spaces):
            print("✅ Aucun espace indésirable dans les noms de colonnes.")

        # --- 3️⃣ Vérification du nombre de colonnes par ligne ---
        anomalies_cols = []
        anomalies_commas = []

        for i, line in enumerate(lines[1:], start=2):
            clean = line.rstrip('\n\r')
            if clean.endswith(sep):
                anomalies_commas.append(i)
                clean = clean[:-1]
            cols = clean.split(sep)
            if len(cols) != nb_cols:
                anomalies_cols.append((i, len(cols)))

        if anomalies_cols:
            print(f"⚠️ {len(anomalies_cols)} lignes avec un nombre de colonnes incorrect.")
        else:
            print("✅ Nombre de colonnes constant sur toutes les lignes.")

        if anomalies_commas:
            print(f"⚠️ Virgule finale détectée sur {len(anomalies_commas)} lignes.")
        else:
            print("✅ Aucune virgule finale trouvée.")

        # --- 4️⃣ Vérification via pandas ---
        df = pd.read_csv(filename, sep=sep)
        unnamed = [col for col in df.columns if "Unnamed" in col]
        if unnamed:
            print(f"⚠️ Colonnes sans nom détectées : {unnamed}")
        else:
            print("✅ Toutes les colonnes ont un nom valide.")

        print("\n🎯 Vérification terminée.\n")

    except FileNotFoundError:
        print(f"❌ Fichier introuvable : {filename}")
    except Exception as e:
        print(f"⚠️ Erreur : {e}")

# --- Exécution ---
verifier_conformite_csv(filename)
```
