```python
import os
import pandas as pd

# === 1️⃣ Dossier courant ===
scenario_path = os.getcwd()
print(f"\n📂 Analyse du dossier courant : {scenario_path}\n")

# === 2️⃣ Fichiers attendus ===
expected_files = [
    "simulation_scenario.csv"
]

# === 3️⃣ Fonction de vérification d’un fichier CSV ===
def analyze_csv(file_path):
    print(f"📄 Analyse de {os.path.basename(file_path)}")
    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print(f"❌ Erreur de lecture : {e}\n")
        return
    
    # Vérification colonnes
    print(f" → Colonnes : {list(df.columns)}")
    cols_with_spaces = [c for c in df.columns if " " in c]
    if cols_with_spaces:
        print(f" ⚠️ Espaces dans les noms de colonnes : {cols_with_spaces}")

    # Vérifier doublons de noms de colonnes
    if len(df.columns) != len(set(df.columns)):
        print(" ⚠️ Colonnes dupliquées détectées")

    # Types de données
    print(" → Types de colonnes :")
    print(df.dtypes)

    # Valeurs manquantes
    missing = df.isna().sum()
    if missing.any():
        print(" ⚠️ Valeurs manquantes détectées :")
        print(missing[missing > 0])

    # Doublons
    if df.duplicated().any():
        print(f" ⚠️ {df.duplicated().sum()} doublons détectés")

    # Taille du fichier
    print(f" → {len(df)} lignes × {len(df.columns)} colonnes\n")

# === 4️⃣ Analyse de tous les fichiers attendus ===
found, missing = 0, 0

for file in expected_files:
    file_path = os.path.join(scenario_path, file)
    if not os.path.exists(file_path):
        print(f"❌ Fichier manquant : {file}")
        missing += 1
    else:
        found += 1
        analyze_csv(file_path)

# === 5️⃣ Rapport final ===
print("\n📊 RAPPORT FINAL")
print(f"✅ Fichiers trouvés : {found}")
print(f"❌ Fichiers manquants : {missing}")
print("✅ Vérification terminée.\n")
```
