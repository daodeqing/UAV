```python
import os
import csv

# === Nom du fichier original et fichier corrigé ===
filename = "olsr_node_state.csv"
output_file = "olsr_node_state_corrige.csv"

def verifier_et_corriger_csv(filename, output_file):
    print(f"🔍 Vérification et correction du fichier : {filename}\n")

    if not os.path.exists(filename):
        print(f"❌ Fichier introuvable : {filename}")
        return

    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    if not lines:
        print("❌ Fichier vide — aucune correction possible.")
        return

    # === 1️⃣ Détection du séparateur ===
    first_line = lines[0].strip()
    sep = ',' if first_line.count(',') >= first_line.count(';') else ';'
    print(f"→ Séparateur détecté : '{sep}'")

    # === 2️⃣ Lecture et nettoyage de l'en-tête ===
    headers_raw = first_line.split(sep)
    headers = [h.strip() for h in headers_raw]
    nb_cols = len(headers)

    print("\n🧾 Colonnes détectées :")
    for h in headers:
        print(f"   - {h}")

    # Détection d'espaces indésirables
    bad_headers = [h for h in headers if h != h.strip() or " " in h]
    if bad_headers:
        print(f"\n⚠️ Espaces détectés dans les noms de colonnes : {bad_headers}")
    else:
        print("\n✅ Aucun espace suspect dans les noms de colonnes.")

    # Correction des noms de colonnes (sans altérer le sens)
    headers_clean = [h.strip().replace(" ", "_") for h in headers]

    # === 3️⃣ Correction ligne par ligne ===
    corrected_lines = [sep.join(headers_clean)]
    nb_corrected = 0

    for i, line in enumerate(lines[1:], start=2):
        clean = line.strip()

        # Supprimer séparateur final inutile
        if clean.endswith(sep):
            clean = clean[:-1]
            nb_corrected += 1

        parts = [p.strip() for p in clean.split(sep)]

        # Ajuster le nombre de colonnes
        if len(parts) < nb_cols:
            parts += [''] * (nb_cols - len(parts))
            nb_corrected += 1
        elif len(parts) > nb_cols:
            parts = parts[:nb_cols]
            nb_corrected += 1

        corrected_lines.append(sep.join(parts))

    # === 4️⃣ Sauvegarde du fichier corrigé ===
    with open(output_file, 'w', encoding='utf-8', newline='') as f:
        for line in corrected_lines:
            f.write(line + '\n')

    # === 5️⃣ Rapport final ===
    print("\n📊 RAPPORT FINAL")
    print(f"✅ Colonnes totales : {nb_cols}")
    print(f"🧩 Lignes traitées  : {len(corrected_lines)-1}")
    print(f"🔧 Lignes corrigées : {nb_corrected}")
    print(f"💾 Fichier corrigé  : {output_file}")
    print("✅ Vérification et correction terminées.\n")


# --- Exécution ---
verifier_et_corriger_csv(filename, output_file)
```
